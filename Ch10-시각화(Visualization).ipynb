## pyplot의 plot() 함수를 이용해 선 그래프 그리기

```python
import numpy as np
import pandas as pd

x_value = [1,2,3,4]
y_value = [2,4,6,8]

plt.plot(x_value, y_value)
plt.show()
```

```python
plt.plot(x_value, y_value, color='green')
```

```python
plt.plot(x_value, y_value, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```

```python
plt.plot(x_value, y_value, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12)

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x_value = np.arange(0, 100)

y_value = 2*x_value

plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xticks(rotation=90)
plt.show()
```

```python
x_value = np.arange(0, 100)

y_value = 2*x_value

plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xticks(ticks=np.arange(0, 100, 5), rotation=90)
plt.title('Hello plot')
plt.show()
```

```python
x_value_1 = np.arange(0, 100)
y_value_1 = 2*x_value_1
y_value_2 = 4*x_value_1

plt.plot(x_value_1, y_value_1, color='green', label='green line')
plt.plot(x_value_1, y_value_2, color='red', label='red line')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()

plt.title('Hello plot')

plt.show()
```

```python
x_value = np.arange(1,10)
y_value = 2*x_value

plt.plot(x_value, y_value, color='red', marker='o', linestyle='dashed', label='red line')

plt.bar(x_value, y_value, color='green', label='bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.title('hello plot')
plt.show()
```

## 여러개의 subplots들을 이용해 개별 그래프들을 subplot별로 시각화하기

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))
```

```python
print('ax type: ', type(ax))
print('ax[0] type:', type(ax[0]))
print('ax[1] type:', type(ax[1]))
```

```python
# 여러개의 subplots 사용시 개별 subplot을 각각 적용해야함
import numpy as np

x_value_1 = np.arange(1, 10)
x_value_2 = np.arange(1, 20)
y_value_1 = 2 * x_value_1
y_value_2 = 2 * x_value_2

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))

ax[0].plot(x_value_1, y_value_1, color='red', marker='o', linestyle='dashed', label='red line')
ax[1].bar(x_value_2, y_value_2, color='green', label='bar plot')

ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')

ax[0].legend()
ax[1].legend()

ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')
plt.show()
```

```python
import numpy as np

x_value_1 = np.arange(1, 10)
x_value_2 = np.arange(1, 20)
y_value_1 = 2 * x_value_1
y_value_2 = 2 * x_value_2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

ax[0][0].plot(x_value_1, y_value_1, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_2, y_value_2, color='green', label='green bar')

ax[1][0].plot(x_value_1, y_value_1, color='green', marker='o', linestyle='dashed', label='green line')
ax[1][1].bar(x_value_2, y_value_2, color='red', label='red bar')

plt.show()
```

## 시본(Seaborn)

### 히스토그램(Histogram)

```python
import pandas as pd

titanic_df = pd.read_csv('train.csv')
titanic_df.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 3))

sns.histplot(titanic_df['Age'], bins=20)
plt.show()
```

```python
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.show()
```

```python
import seaborn as sns

# seaborn의 figure레벨 그래프는 p[t.figure()로 Figure 크기를 조절할 수 없습니다.
plt.figure(figsize=(8, 4))
sns.displot(x='Age', data=titanic_df, kde=True)
plt.show()
```

### subplots를 이용하여 시본의 다양한 그래프를 시각화

```python
cat_columns = ['Survived', 'Pclass', 'Sex']

fig, axs = plt.subplots(nrows=1, ncols=len(cat_columns), figsize=(14,4))

for index, column in enumerate(cat_columns):
    print('index: ', index)
    sns.countplot(x=column, data=titanic_df, ax=axs[index])
```

```python
cat_columns = ['Pclass', 'Sex', 'Embarked']

fig, axs = plt.subplots(nrows=1, ncols=len(cat_columns), figsize=(14,4))

for index, column in enumerate(cat_columns):
    print('index: ', index)
    sns.barplot(x=column, y='Survived', data=titanic_df, ax=axs[index])
```

```python
cont_columns = ['Age', 'Fare']

for column in cont_columns:
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
    sns.violinplot(x='Survived', y=column, data=titanic_df, ax=axs[0])
    sns.histplot(x=column, ndata=titanic_df, kde=True, hue='Survived', ax=axs[1])
```
